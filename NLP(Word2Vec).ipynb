```python
from gensim.models import Word2Vec
# Take user input for sentences
n = int(input("Enter number of sentences: "))
sentences = []
for i in range(n):
  line = input(f"Enter sentence {i+1}: ").lower().split()
  sentences.append(line)

# Choose model type
choice = input("\nChoose model type (cbow/skipgram): ").strip().lower()

if choice == "cbow":
  model = Word2Vec(sentences, vector_size=50, window=3, min_count=1, sg=0)
  print("\nCBOW model trained successfully!")
elif choice == "skipgram":
  model = Word2Vec(sentences, vector_size=50, window=3, min_count=1, sg=1)
  print("\nSkip-gram model trained successfully!")
else:
  print("Invalid choice! Please enter 'cbow' or 'skipgram'.")
  exit()
# Ask for word
word = input("\nEnter a word to check embedding and similar words: ").lower()
# Display results
if word in model.wv:
  print(f"\nEmbedding for '{word}':\n", model.wv[word])
  print(f"\nMost similar to '{word}':", model.wv.most_similar(word))

else:
  print(f"\nWord '{word}' not in vocabulary.")
```
